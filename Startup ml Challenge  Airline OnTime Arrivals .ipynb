```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 

#Array processing
import numpy as np
#Data analysis, wrangling and common exploratory operations
import pandas as pd
from pandas import Series, DataFrame
from sklearn.neighbors import KNeighborsClassifier    
#For visualization. Matplotlib for basic viz and seaborn for more stylish figures + statistical figures not in MPL.
import matplotlib.pyplot as plt
import seaborn as sns

#For some of the date operations
import datetime
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor 
from sklearn.svm import SVC
from sklearn.linear_model import  LogisticRegression, LinearRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,MultinomialNB
```

```python
fec_all = pd.read_csv('178329925_T_ONTIME.csv',usecols=['ORIGIN_AIRPORT_ID','DEST_AIRPORT_ID','DEP_TIME','CRS_DEP_TIME','CRS_ARR_TIME'],low_memory=False)
lables = pd.read_csv('178329925_T_ONTIME.csv',usecols=['ARR_DEL15'])
replace = fec_all.fillna(0)
lables = lables.fillna(0)
```

```python
traing_data = replace[:150000]
traing_labels = lables[:150000]
test_data  = replace[150000:]
test_labels =np.array(lables[150000:])
```

```python
test_data
```

```python
  
cart_model_default = DecisionTreeClassifier(random_state = 1234, criterion = "entropy") 
cart_model_default.fit(traing_data,traing_labels)
print(cart_model_default) 
predictions_dtree_default = cart_model_default.predict(test_data)
print (metrics.accuracy_score(test_labels,predictions_dtree_default) )
```

```python
nb_gaussian_model = GaussianNB() #create the model
nb_gaussian_model.fit(traing_data,traing_labels)
print(nb_gaussian_model) 
predictions_gaussian_nb = nb_gaussian_model.predict(test_data) #make the predictions
print (metrics.accuracy_score(test_labels,predictions_gaussian_nb) )
```

```python
lr_model_default = LogisticRegression(random_state = 1234)      
lr_model_default.fit(traing_data, traing_labels)
predictions_lr_model_default = lr_model_default.predict(test_data)
print(metrics.accuracy_score(test_labels,predictions_lr_model_default))
```

```python
rf_model_default = RandomForestClassifier(random_state = 1234)                                                                    
rf_model_default.fit(traing_data, traing_labels)                                                                    
print(rf_model_default)            
predictions_rf_default = rf_model_default.predict(test_data)
print (metrics.accuracy_score(test_labels,predictions_rf_default) )
```
